```python
import opendatasets as od
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib
import os
```

```python
od.download('https://www.piworks.net/Upload/Document/Original/country_vaccination_stats.csv')
```

```python
raw_df = pd.read_csv('country_vaccination_stats.csv')
```

```python
raw_df
```

```python
raw_df['daily_vaccinations'] = raw_df.groupby('country')['daily_vaccinations'].transform(lambda x:x.fillna(x.min()))
```

```python
raw_df['daily_vaccinations'].fillna(0,inplace=True)
```

```python
countryVaccMedian = raw_df.groupby('country')['daily_vaccinations'].median().sort_values(ascending=False).head(3)
countryVaccMedian
```
